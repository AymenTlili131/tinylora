```python
import os, glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
src = "results/summary_tinylora.csv"
dst = "results/summary_tinylora_fp16_steps200_r2.csv"

if os.path.exists(src):
    if not os.path.exists(dst):
        os.rename(src, dst)
        print("Renamed:", src, "->", dst)
    else:
        print("Destination already exists:", dst)
else:
    print("Source CSV not found:", src)
```

```python
# Load any summary_*.csv under results/ (including subfolders)
paths = glob.glob("results/**/summary_*.csv", recursive=True)

dfs = []
for p in paths:
    try:
        df = pd.read_csv(p)
        df["source_csv"] = p
        dfs.append(df)
    except Exception as e:
        print("Skip", p, ":", e)

all_df = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()
all_df
```

```python
if all_df.empty:
    raise RuntimeError("No CSVs loaded. Check results/ paths.")

# If multiple runs exist, we pick the row from the lexicographically latest source_csv
# (works well if you use unique filenames / folders).
all_df = all_df.sort_values("source_csv").groupby(["model", "method", "task"], as_index=False).tail(1)

# Ensure numeric columns
for c in ["trainable_params", "peak_vram_gb", "train_seconds", "tokens_per_sec_est", "steps"]:
    if c in all_df.columns:
        all_df[c] = pd.to_numeric(all_df[c], errors="coerce")

all_df.sort_values(["task","method"])
```

```python
def score_column_for_task(task: str):
    task = task.lower()
    if task == "math":
        return "math_exact_match"
    if task == "cls":
        return "cls_accuracy"
    if task == "instruct":
        return "rougeL"
    return None
```

```python
def plot_task(df_task: pd.DataFrame, task: str, title_suffix=""):
    score_col = score_column_for_task(task)
    if score_col is None or score_col not in df_task.columns:
        print(f"Missing score column for task={task}: {score_col}")
        return

    d = df_task.dropna(subset=["trainable_params", score_col, "peak_vram_gb"]).copy()
    if d.empty:
        print("No data for task:", task)
        return

    # Bubble size scaling (VRAM)
    v = d["peak_vram_gb"].to_numpy()
    # Avoid giant bubbles
    sizes = 200 * (v / max(v.max(), 1e-9))**1.2 + 40

    plt.figure(figsize=(8, 5))
    plt.scatter(d["trainable_params"], d[score_col], s=sizes, alpha=0.7)

    for _, r in d.iterrows():
        plt.annotate(r["method"], (r["trainable_params"], r[score_col]),
                     textcoords="offset points", xytext=(6, 4), fontsize=9)

    plt.xscale("log")
    plt.xlabel("Trainable parameters (log scale)")
    plt.ylabel(score_col)
    plt.title(f"{task.upper()} — params vs score (bubble=VRAM GB){title_suffix}")
    plt.grid(True, which="both", linestyle="--", alpha=0.3)
    plt.show()

for task in ["math", "cls", "instruct"]:
    plot_task(all_df[all_df["task"] == task], task)
```

```python
cols = ["task","method","trainable_params","peak_vram_gb","train_seconds","tokens_per_sec_est","steps",
        "math_exact_match","cls_accuracy","rougeL","source_csv"]
present = [c for c in cols if c in all_df.columns]
all_df[present].sort_values(["task","trainable_params"])
```

```python

def plot_params_vs_score(df, task, score_col):
    d = df[df["task"]==task].copy()
    d = d.dropna(subset=[score_col, "trainable_params", "peak_vram_gb"])

    fig = plt.figure(figsize=(10,4))
    ax = plt.gca()

    ax.scatter(d["trainable_params"], d[score_col])  # no bubble sizes
    ax.set_xscale("log")
    ax.set_xlabel("Trainable parameters (log scale)")
    ax.set_ylabel(score_col)
    ax.set_title(f"{task.upper()} — params vs {score_col}")

    for _, r in d.iterrows():
        label = f'{r["method"]}\npeak={r["peak_vram_gb"]:.2f}GB'
        # if you added inference_steady_gb:
        if "inference_steady_gb" in d.columns and not (r.get("inference_steady_gb") is None):
            label += f'\nsteady={r["inference_steady_gb"]:.2f}GB'
        ax.annotate(label, (r["trainable_params"], r[score_col]), textcoords="offset points", xytext=(5,5))

    plt.show()

# Examples:
plot_params_vs_score(all_df, "math", "math_exact_match")
plot_params_vs_score(all_df, "cls", "cls_accuracy")
plot_params_vs_score(all_df, "instruct", "rougeL")
```

```python
import os, glob
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from data import load_task_dataset
from evals import generate_batch

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# 1) Show one example from each dataset
for task in ["math", "cls", "instruct"]:
    ds = load_task_dataset(task, split="train", limit=5)
    ex = ds[0]
    print("\n" + "="*80)
    print(f"TASK: {task.upper()}")
    print("- Prompt:")
    print(ex["prompt"][:800] + ("..." if len(ex["prompt"])>800 else ""))
    print("- Target:")
    print(ex["response"][:400] + ("..." if len(ex["response"])>400 else ""))

# 2) Helper: pick latest checkpoint folder for a given task+method
def latest_run_dir(task: str, method: str, outdir="results"):
    # run.py uses output_dir = results/{task}_{method}
    base = os.path.join(outdir, f"{task}_{method}")
    if not os.path.isdir(base):
        return None
    # Trainer saves checkpoints or final model in that folder; pick the folder itself
    return base

# 3) Load models for a given task (you can change TASK below)
TASK = "math"   # try: "cls" or "instruct"

methods = ["tinylora", "lora", "adalora", "vera", "vblora"]
ckpts = {m: latest_run_dir(TASK, m) for m in methods}
print("\nCheckpoint dirs:", ckpts)

# 4) Load tokenizer once (same base model)
base_model_name = "Qwen/Qwen2.5-0.5B-Instruct"
tok = AutoTokenizer.from_pretrained(base_model_name, use_fast=True)
if tok.pad_token is None:
    tok.pad_token = tok.eos_token

def load_ft_model(run_dir: str):
    # Loads whatever Trainer saved in that run_dir
    model = AutoModelForCausalLM.from_pretrained(
        run_dir,
        torch_dtype=torch.float16 if DEVICE=="cuda" else torch.float32,
        device_map="auto" if DEVICE=="cuda" else None,
    )
    model.eval()
    return model

loaded = {}
for m, d in ckpts.items():
    if d is None:
        continue
    try:
        loaded[m] = load_ft_model(d)
    except Exception as e:
        print(f"Could not load {m} from {d}: {e}")

# 5) Choose a prompt from dataset + run all models
demo_ds = load_task_dataset(TASK, split="train", limit=50)
idx = 0  # change this index to pick a different example
prompt = demo_ds[idx]["prompt"]

print("\n" + "="*80)
print("DEMO PROMPT:\n", prompt[:1200])

for m, model in loaded.items():
    out = generate_batch(model, tok, [prompt], max_new_tokens=128, temperature=0.0)[0]
    print("\n" + "-"*80)
    print(f"METHOD: {m}")
    print(out[:1500])
```
